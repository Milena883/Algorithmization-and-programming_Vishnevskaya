**Задача 1**: Построение точек динозавра: создайте Python-программу для визуализации динозавра, используя точки и векторы.

Input

```python
from matplotlib import pyplot as plt#Здесь происходит импорт подмодуля pyplot из библиотеки matplotlib, который предоставляет функции для построения графиков.


# Список векторов, определяющих динозавра

dino_vectors = [(6, 4), (3, 1), (1, 2), (-1, 5), (-2, 5), (-3, 4), (-4, 4),

                (-5, 3), (-5, 2), (-2, 2), (-5, 1), (-4, 0), (-2, 1),

                (-1, 0), (0, -3), (-1, -4), (1, -4), (2, -3), (1, -2),

                (3, -1), (5, 1)]#В этом списке определяются координаты (x, y) точек, которые формируют контур динозавра. Каждая пара чисел соответствует одной точке на графике.


# Разделяем координаты x и y

x_coords, y_coords = zip(*dino_vectors)#Метод zip(*...) используется для распаковки списка кортежей и разделения их на два отдельных списка: один для координат x, другой для координат y.


# Построение графика

plt.figure(figsize=(8, 6))#Создается новая фигура для графика с заданными размерами 8 на 6 дюймов.

plt.scatter(x_coords, y_coords, color="orange", label="Точки динозавра")#Используется метод scatter для построения рассеянного графика (точечного графика),

plt.title("Точки контура динозавра")#title устанавливает заголовок графика.

plt.xlabel("X-координаты")#xlabel и ylabel задают подписи для осей X и Y соответственно.

plt.ylabel("Y-координаты")#xlabel и ylabel задают подписи для осей X и Y соответственно.

plt.grid(True)#Включает сетку на графике для лучшего восприятия.

plt.legend()#Добавляет легенду, которая отображает метку "Точки динозавра".

plt.show()#Этот метод выводит график на экран.
```

Вывод: на графике отобразятся только точки, составляющие контур динозавра.

Вращение динозавра

Input

```python
from math import radians, cos, sin#Здесь импортируются функции radians, cos и sin из модуля math.


# Функция вращения

def rotate(vector, theta):#Функция rotate принимает два аргумента: vector (координаты точки в виде кортежа) и theta (угол вращения в градусах).

    angle = radians(theta)

    return (vector[0] * cos(angle) - vector[1] * sin(angle),

            vector[0] * sin(angle) + vector[1] * cos(angle))


# Вращение фигуры

theta = 45

rotated_dino = [rotate(v, theta) for v in dino_vectors]
#Здесь задается угол вращения theta, равный 45 градусам.
#Используется генератор списка, который применяет функцию rotate ко всем точкам в списке dino_vectors.
#Для каждой точки вычисляются новые координаты после вращения на 45 градусов, и результат сохраняется в переменной rotated_dino.

# Визуализация

plt.figure(figsize=(8, 6))

plt.plot(*zip(*dino_vectors), linestyle="-", color="grey", label="Исходный динозавр")

plt.plot(*zip(*rotated_dino), linestyle="--", color="purple", label=f"Динозавр, вращённый на {theta}°")

plt.title("Вращение динозавра")

plt.legend()

plt.grid(True)

plt.show()
```

**Задача 4:**  Перемещение динозавра Смоделируйте сдвиг фигуры динозавра на ( − 2 , − 2 ) (−2,−2).

Input

```python
def shift_vectors(vectors, dx, dy):

    return [(x + dx, y + dy) for x, y in vectors]


shifted_dino = shift_vectors(dino_vectors, -2, -2)


# Отображение исходного и сдвинутого динозавров

plt.figure(figsize=(8, 6))

plt.plot(*zip(*dino_vectors), linestyle="-", color="blue", label="Исходный динозавр")

plt.plot(*zip(*shifted_dino), linestyle="--", color="orange", label="Сдвинутый динозавр")

plt.scatter(*zip(*dino_vectors), color="blue")

plt.scatter(*zip(*shifted_dino), color="orange")

plt.title("Перемещение динозавра")

plt.xlabel("X")

plt.ylabel("Y")

plt.grid(True)

plt.legend()

plt.show()
```

**Задача 5:** Построение сотни динозавров

Создайте множество копий динозавра, расположенных в сетке.

Input

```python
def create_dino_grid(vectors, grid_size=5, spacing=12):

    dinos = []

    for i in range(-grid_size, grid_size):

        for j in range(-grid_size, grid_size):

            dinos.append(shift_vectors(vectors, i * spacing, j * spacing))

    return dinos


grid_dinos = create_dino_grid(dino_vectors)


# Построение всех динозавров

plt.figure(figsize=(10, 10))

for dino in grid_dinos:

    plt.plot(*zip(*dino), linestyle="-", color="blue")


plt.title("Сетка из динозавров")

plt.xlabel("X")

plt.ylabel("Y")

plt.grid(False)

plt.show()
```

Вывод: на графике будет отображено множество фигур динозавра.

**Задача 6:** сложение векторов Используя функцию add, напишите программу для визуализации суммы двух векторов (4 , 3 ) (4,3) и ( − 1 , 1 ) (−1,1).

Input

```python
# Функция сложения векторов

def add(v1, v2):

    return (v1[0] + v2[0], v1[1] + v2[1])


# Исходные векторы

v1 = (4, 3)

v2 = (-1, 1)

v_sum = add(v1, v2)


# Построение векторов

plt.figure(figsize=(6, 6))

plt.quiver(0, 0, *v1, angles="xy", scale_units="xy", scale=1, color="blue", label="Вектор v1 (4, 3)")

plt.quiver(0, 0, *v2, angles="xy", scale_units="xy", scale=1, color="green", label="Вектор v2 (-1, 1)")

plt.quiver(0, 0, *v_sum, angles="xy", scale_units="xy", scale=1, color="red", label="Сумма v1 + v2 (3, 4)")


# Настройка осей

plt.xlim(-2, 5)

plt.ylim(-2, 5)

plt.axhline(0, color="black",linewidth=0.5)

plt.axvline(0, color="black",linewidth=0.5)

plt.grid(True)

plt.gca().set_aspect('equal', adjustable='box')

plt.legend()

plt.title("Сложение векторов")

plt.show()
```

Вывод: Визуализация двух исходных векторов и их суммы.




**Задача 7:** Умножение вектора на скаляр

Реализуйте умножение вектора 𝑣 = ( 2 , 3 ) v=(2,3) на скаляр 𝑠 = 2.5 s=2.5 и визуализируйте его.

Input

```python
# Функция скалярного умножения

def scale(scalar, v):

    return (scalar * v[0], scalar * v[1])


# Исходный вектор

v = (2, 3)

scaled_v = scale(2.5, v)


# Построение векторов

plt.figure(figsize=(6, 6))

plt.quiver(0, 0, *v, angles="xy", scale_units="xy", scale=1, color="blue", label="Вектор v (2, 3)")

plt.quiver(0, 0, *scaled_v, angles="xy", scale_units="xy", scale=1, color="red", label="Скалярное произведение 2.5v (5, 7.5)")


# Настройка осей

plt.xlim(-1, 6)

plt.ylim(-1, 8)

plt.axhline(0, color="black",linewidth=0.5)

plt.axvline(0, color="black",linewidth=0.5)

plt.grid(True)

plt.gca().set_aspect('equal', adjustable='box')

plt.legend()

plt.title("Умножение вектора на скаляр")

plt.show()
```

Вывод: График показывает исходный вектор и результат его масштабирования.



**Задача 8:**Построение длины вектора

Реализуйте вычисление длины вектора 𝑣 = ( 4 , 3 ) v=(4,3) с использованием теоремы Пифагора.

Input

```python
from math import sqrt


# Функция для вычисления длины вектора

def length(v):

    return sqrt(v[0]**2 + v[1]**2)


# Исходный вектор

v = (4, 3)

v_len = length(v)


print(f"Длина вектора {v}: {v_len:.2f}")
```

**Задача 9**: Смещение точки

Создайте функцию для смещения точки (x,y) на заданный вектор (dx,dy). Например, точки (1,2) на (-2,-1)



Input



```python
# Функция для смещения точки

def translate(translation, points):

    return [(p[0] + translation[0], p[1] + translation[1]) for p in points]


# Исходные точки

points = [(1, 2), (3, 4), (5, 6)]

translation = (-2, -1)

shifted_points = translate(translation, points)


# Вывод результата

print(f"Исходные точки: {points}")

print(f"Смещённые точки: {shifted_points}")
```

Output



Исходные точки: [(1, 2), (3, 4), (5, 6)]

Смещённые точки: [(-1, 1), (1, 3), (3, 5)]



Вывод: Список новых координат точек после сдвига.

**Задача 10**: Построение параллелограмма



Нарисуйте параллелограмм, построенный по двум векторам u=(3, 20) и v=(1, 4).

Input

```python
# Функция построения параллелограмма

def parallelogram(v1, v2):#

    return [v1, add(v1, v2), v2, (0, 0), v1]


# Исходные векторы

u = (3, 2)

v = (1, 4)

parallelogram_points = parallelogram(u, v)#


# Построение

plt.figure(figsize=(6, 6))#

plt.plot(*zip(*parallelogram_points), color="orange", linestyle="-", marker="o", label="Параллелограмм")#

plt.quiver(0, 0, *u, angles="xy", scale_units="xy", scale=1, color="blue", label="Вектор u")#

plt.quiver(0, 0, *v, angles="xy", scale_units="xy", scale=1, color="green", label="Вектор v")#


# Настройка осей

plt.xlim(-1, 6)

plt.ylim(-1, 6)

plt.axhline(0, color="black",linewidth=0.5)

plt.axvline(0, color="black",linewidth=0.5)

plt.grid(True)

plt.gca().set_aspect('equal', adjustable='box')

plt.legend()

plt.title("Параллелограмм по двум векторам")

plt.show()
```

**Задача 11:** Вычитание векторов

 Создайте программу для вычисления двух векторов v1=(5, 4) и v2=2, 3.

Отобразите векторы и их разность.

Input

```python
# Функция для вычитания векторов

def subtract(v1, v2):

    return (v1[0] - v2[0], v1[1] - v2[1])


# Исходные векторы

v1 = (5, 4)

v2 = (2, 3)

v_diff = subtract(v1, v2)


# Визуализация

plt.figure(figsize=(6, 6))

plt.quiver(0, 0, *v1, angles="xy", scale_units="xy", scale=1, color="blue", label="Вектор v1 (5, 4)")

plt.quiver(0, 0, *v2, angles="xy", scale_units="xy", scale=1, color="green", label="Вектор v2 (2, 3)")

plt.quiver(0, 0, *v_diff, angles="xy", scale_units="xy", scale=1, color="red", label="Разность v1 - v2 (3, 1)")


# Настройка осей

plt.xlim(-1, 6)

plt.ylim(-1, 6)

plt.axhline(0, color="black", linewidth=0.5)

plt.axvline(0, color="black", linewidth=0.5)

plt.grid(True)

plt.gca().set_aspect('equal', adjustable='box')

plt.legend()

plt.title("Вычитание векторов")

plt.show()
```

Вывод: График, отображающий два исходных вектора и их разность.

**Задача 12:** Расстояние между двумя точками

Напишите программу для вычисления расстояния между двумя точками p1=(1, 2) и p2=(4, 6), используя теорему Пифагора.

Input

```python

# Функция для вычисления расстояния

def distance(p1, p2):

    return sqrt((p2[0] - p1[0])**2 + (p2[1] - p1[1])**2)


# Исходные точки

p1 = (1, 2)

p2 = (4, 6)

dist = distance(p1, p2)


print(f"Расстояние между точками {p1} и {p2}: {dist:.2f}")
```

Output

Расстояние между точками (1, 2) и (4, 6): 5.00
