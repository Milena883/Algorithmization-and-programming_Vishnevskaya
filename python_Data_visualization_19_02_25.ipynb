# **Построение графика**

Для начала построим простую линейную зависимость, зададим название графику, подпишем оси и отобразим сетку:



```python
import numpy as np
import matplotlib.pyplot as plt

# Независимая (x) и зависимая (y) переменные
x = np.linspace(0, 10, 50)
y = x  # Исправлено с 'у' на 'y'

# Построение графика
plt.title('Линейная зависимость y = x')  # заголовок
plt.xlabel('x')  # ось абсцисс
plt.ylabel('y')  # ось ординат
plt.grid()  # включение отображения сетки
plt.plot(x, y)  # построение графика
plt.show()  # отображаем график
```

Таким образом, код эффективно демонстрирует линейную зависимость y=x в интервале от 0 до 10, создавая простую и понятную визуализацию.

Изменим тип линии и её цвет, для этого в функцию plot(), в качестве третьего параметра, передадим строку, сформированную определенным образом, в нашем случае это 'г--', где 'г' означает красный цвет, а '--' - тип линии — пунктирная линия:

# **Несколько графиков на одном поле**


Построим несколько графиков на одном поле, для этого добавим
квадратичную зависимость:




```python
import numpy as np
import matplotlib.pyplot as plt

# Задаем независимую (x) и зависимую (y) переменные
x = np.linspace(0, 1, 10)
y = x  # Зависимая переменная y равна x

# Построение графика
plt.title('Линейная зависимость y = x')  # заголовок
plt.xlabel('x')  # ось абсцисс
plt.ylabel('y')  # ось ординат
plt.grid()  # включение отображения сетки
plt.plot(x, y, 'r--')  # построение графика с красной пунктирной линией
plt.show()  # отображаем график
```

Ваш код правильно создает и отображает линейный график зависимости y=x в диапазоне от 0 до 1 с использованием красной пунктирной линии. График легко интерпретировать благодаря четким меткам осей и заголовку.

# **Несколько графиков на одном поле**

Построим несколько графиков на одном поле, для этого добавим
квадратичную зависимость:


```python
import numpy as np
import matplotlib.pyplot as plt

# Линейная зависимость
x = np.linspace(0, 20, 5)
y1 = x

# Квадратичная зависимость
y2 = [i**2 for i in x]

# Построение графика
plt.title('Зависимости: y1 = x, y2 = x^2')  # исправлено 'хл2' на 'x^2'
plt.xlabel('x')  # ось абсцисс
plt.ylabel('y1, y2')  # ось ординат
plt.grid()  # включение отображения сетки
plt.plot(x, y1, label='y1 = x')  # линейная зависимость
plt.plot(x, y2, label='y2 = x^2')  # квадратичная зависимость
plt.legend()  # добавляет легенду
plt.show()  # отображаем график
```

Ваш код эффективно демонстрирует линейные и квадратичные зависимости между переменной
x и значениями y1 и y2 на графике. Легенда и сетка делают график более информативным и легким для интерпретации.

# **Построение диаграммы для категориальных данных**


До этого мы строили графики по численным данным, то есть зависимая
и независимая переменные имели числовой тип. На практике довольно

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из CSV файла
def load_data(file_path):
    try:
        data = pd.read_csv(file_path)
        print("Данные успешно загружены.")
        return data
    except FileNotFoundError:
        print("Файл не найден. Проверьте путь к файлу.")
        return None

# Обработка данных: вычисление среднего значения
def process_data(data):
    mean_score = data['score'].mean()
    print(f"Средний балл: {mean_score:.2f}")
    return mean_score

# Визуализация данных
def visualize_data(data):
    plt.figure(figsize=(10, 6))
    plt.bar(data['name'], data['score'], color='skyblue')
    plt.xlabel('Имя')
    plt.ylabel('Баллы')
    plt.title('Баллы студентов')
    plt.axhline(y=data['score'].mean(), color='r', linestyle='--', label='Средний балл')
    plt.legend()
    plt.show()

# Основная функция
def main():
    file_path = 'data.csv'  # Путь к файлу с данными
    data = load_data(file_path)

    if data is not None:
        process_data(data)
        visualize_data(data)

if __name__ == "__main__":
    main()
```
